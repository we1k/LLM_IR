# PDF文件的处理
## pdf2html

* The simplest case
Suppose you have a PDF file pdf/test.pdf, simply running

pdf2htmlEX --zoom 1.3 pdf/test.pdf

would produce a single HTML file test.html in the current directory.

* Advanced
pdf2htmlEX -f 3 -l 5 --fit-width 1024 --bg-format jpg pdf/test.pdf
would convert only the 3rd, 4th and 5th pages, and fit the page width to 1024 pixels. Background images will be generated in the JPEG format.

* For publishers
pdf2htmlEX --embed cfijo --dest-dir out pdf/test.pdf
would produce a test.html and accompanying files in the out directory, in this way all the resources (fonts, images, css and javascript) are stored in separated files such that the viewer can take more advantage of browser caches.

## html 划分标题
先找大标题和小标题
```python
# ff6 fs0 fc0 sc0 ls0 ws0 方向盘加热（大标题）
# ff1 fs0 fc0 sc0 ls0 ws0 后排头枕（小标题）
# ff2 fs5 fc3 sc0 ls0 ws0 警告！
# ff2 fs5 fc4 sc0 ls0 ws0 注意！
# h4 y9 ff4 fs0 fc0 sc0 ls0 ws0 页码
# d m1 页眉
``` 

````python
# 删除页眉和页码
filepath = "out/初赛训练数据集.html"
finalpath = "out/final.html"
from bs4 import BeautifulSoup

# 读取HTML文件
with open(filepath, 'r', encoding='utf-8') as file:
    content = file.read()

# 使用BeautifulSoup解析HTML
soup = BeautifulSoup(content, 'lxml')

# 假设页码、页眉和页脚都在某些特定的标签内，例如<div class="page-number">, <div class="header">, <div class="footer"> 
# 我们可以根据实际情况调整选择器
for tag in soup.find_all(class_="d m1"):  # 链接
    tag.decompose()

for tag in soup.find_all(class_=lambda value: value and value.endswith("h5 ye6 ff1 fs1 fc0 sc0 ls0 ws0")):  # "t m0 xc "左页眉 t m0 x22 右页眉
    tag.decompose()
    
for tag in soup.find_all(class_=lambda value: value and value.endswith("h4 y9 ff4 fs0 fc0 sc0 ls0 ws0")): # 页码
    tag.decompose()

    # "t m0 xd h9 ye4 ff3 fs3 fc2 sc0 ls0 ws0"

# 将处理后的HTML保存到新文件或同一个文件
with open(finalpath, 'w', encoding='utf-8') as file:
    file.write(str(soup))
````

````python
finalpath = "data/out/final2.html"
bigheader = "ff6 fs0 fc0 sc0 ls0 ws0"
smallheader = "ff1 fs0 fc0 sc0 ls0 ws0"
from bs4 import BeautifulSoup

# 读取HTML文件
with open(finalpath, 'r', encoding='utf-8') as file:
    content = file.read()

# 使用BeautifulSoup解析HTML
soup = BeautifulSoup(content, 'lxml')

# 查找所有具有指定class的标签
tags = soup.find_all(class_=lambda value: value and (value.endswith(bigheader) or value.endswith(smallheader)))

# 打印找到的标签的文本内容
for tag in tags:
    print(tag.text)
````

````python
import re
from bs4 import BeautifulSoup

# 读取HTML文件
with open(finalpath, 'r', encoding='utf-8') as file:
    content = file.read()

# 使用BeautifulSoup解析HTML
soup = BeautifulSoup(content, 'lxml')

# 查找所有具有指定class结尾的标签
delimiter_tags = soup.find_all(class_=lambda value: value and (value.endswith(bigheader) or value.endswith(smallheader)))

# 获取所有符合条件的标签的完整HTML，并为正则表达式预处理这些内容
delimiter_patterns = [re.escape(str(tag)) for tag in delimiter_tags]

# 使用找到的标签的HTML作为分隔符，将HTML文本分割成多个部分
split_content = re.split('|'.join(delimiter_patterns), content)

# 为每个分割的部分进一步处理，只保留class以"t m0"开头的部分，并获取其纯文本内容
processed_parts = []
for part in split_content:
    part_soup = BeautifulSoup(part, 'lxml')
    relevant_tags = part_soup.find_all(class_=lambda value: value and value.startswith("t m0"))
    processed_parts.append("".join([tag.get_text(strip=True) for tag in relevant_tags]))
````

````python
import json
data = {}
for key, value in zip(delimiter_tags[1:],processed_parts[2:]):
    data[key.text] = value

# 将字典转换为JSON格式字符串
json_data = json.dumps(data,ensure_ascii=False, indent=4)

print(json_data)
with open("header_all.json", "w", encoding="utf-8") as file:
    file.write(json_data)
````

````python
from modelscope.pipelines import pipeline
# Version less than 1.1 please use TextRankingPreprocessor
from modelscope.utils.constant import Tasks
import json

def get_context(pipeline,question,data) -> str:
    """
    根据问题与索引标题的相似度进行检索，返回最相似的结果
    """
    keylist = list(data.keys())
    # 打印解析后的数据
    inputs = {
        'source_sentence': [question],
        'sentences_to_compare': keylist
    }
    result = pipeline(input=inputs)
    print (result["scores"])
    # 使用max()函数找到列表中的最大值
    max_value = max(result["scores"])
    # 使用index()方法找到最大值的索引
    max_index = result["scores"].index(max_value)

    return data[keylist[max_index]]

model_id = './sbert/damo/chinese/'
pipeline_ins = pipeline(task=Tasks.text_ranking, model=model_id)
# 读取JSON文件
with open("bigheader.json", "r", encoding="utf-8") as file:
    json_data = file.read()
data = json.loads(json_data)
ret = get_context(pipeline_ins,"怎样加热座椅？",data)
print(ret)
````

````python
prompt_template = """根据已知信息筛选出最相关内容并简洁地回答问题。
问题是： {}
已知信息有： {}
"""

question = "怎么打开危险警告灯？"
inputs = {
    "question": question,
    "info":get_context(pipeline_ins,question,data)
    }

# Execute the chain
response, history = model.chat(tokenizer,prompt_template.format(inputs["question"],inputs["info"]), history=[])
print(prompt_template.format(inputs["question"],inputs["info"]))
print(response)
````

````python
import json

# 读取JSON文件
input_file = 'result/answer.json'
output_file = 'result/final03.json'

with open(input_file, 'r') as file:
    data = json.load(file)

# 保留question和answer字段
filtered_data = []
for item in data:
    filtered_item = {
        "question": item.get("question"),
        "answer_1": item.get("answer_0").replace("\n", ""),
        "answer_2": item.get("answer_1").replace("\n", ""),
        "answer_3": ""
    }
    filtered_data.append(filtered_item)

# 写入新的JSON文件
with open(output_file, 'w') as file:
    json.dump(filtered_data, file,ensure_ascii=False,indent=4)

print("JSON file processing complete.")
````
